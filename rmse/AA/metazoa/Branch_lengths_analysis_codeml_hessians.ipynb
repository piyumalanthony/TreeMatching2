```python
import copy
import numpy as np
import matplotlib.pyplot as plt
```

```python
iqtree_file_LG = 'LG_mcmctree_metazoa_out.BV'
iqtree_file_LG_G4 = 'LG_G4_mcmctree_metazoa_out.BV'
```

```python
hessian_list = []

hessians_LG = []
treestrings_LG = []
branch_lengths_LG = []
gradient_vectors_LG = []
counter = 0
with open(iqtree_file_LG) as f:
    for k, line in enumerate(f.readlines()):
        if counter == 3:
            treestrings_LG.append(line)
            print('tree string',k)
        if counter == 5:
            branch_lengths_numerical = line.strip().split()
            branch_lengths_LG.append([float(i) for i in branch_lengths_numerical])
            print('branch lengths',k)
        if counter == 7:
            gradients_numerical = line.strip().split()
            gradient_vectors_LG.append([float(i) for i in gradients_numerical])
            print('gradients',k)
            
        if counter > 11 and line.strip() == '':
            hessians_LG.append(copy.deepcopy(hessian_list))
            hessian_list = []
            counter = 0 
            print('Hessian all string',k)
        elif counter > 11:
            # print(i, line)
            hessian_row_numerical = line.strip().split()
            hessian_list.append([float(i) for i in hessian_row_numerical])
            print('local hessian',k)
        counter += 1
    else:
        hessians_LG.append(copy.deepcopy(hessian_list))
```

```python
branch_lengths_LG = np.array(branch_lengths_LG)
```

```python
branch_lengths_LG.shape
```

```python
branch_lengths_LG = branch_lengths_LG.reshape(-1,1)
```

```python
hessian_list_LG_G4 = []

hessians_LG_G4 = []
treestrings_LG_G4 = []
branch_lengths_LG_G4 = []
gradient_vectors_LG_G4 = []
counter = 0
with open(iqtree_file_LG_G4) as f:
    for k, line in enumerate(f.readlines()):
        if counter == 3:
            treestrings_LG_G4.append(line)
            print('tree string',k)
        if counter == 5:
            branch_lengths_numerical = line.strip().split()
            branch_lengths_LG_G4.append([float(i) for i in branch_lengths_numerical])
            print('branch lengths',k)
        if counter == 7:
            gradients_numerical = line.strip().split()
            gradient_vectors_LG_G4.append([float(i) for i in gradients_numerical])
            print('gradients',k)
            
        if counter > 11 and line.strip() == '':
            hessians_LG_G4.append(copy.deepcopy(hessian_list))
            hessian_list = []
            counter = 0 
            print('Hessian all string',k)
        elif counter > 11:
            # print(i, line)
            hessian_row_numerical = line.strip().split()
            hessian_list_LG_G4.append([float(i) for i in hessian_row_numerical])
            print('local hessian',k)
        counter += 1
    else:
        hessians_LG_G4.append(copy.deepcopy(hessian_list))
```

```python
branch_lengths_LG_G4 = np.array(branch_lengths_LG_G4)
```

```python
branch_lengths_LG_G4.shape
```

```python
branch_lengths_LG_G4 = branch_lengths_LG_G4.reshape(-1,1)
```

```python
branch_lengths_LG_G4
```

```python
fig, ax = plt.subplots(figsize=(9 ,9))
ax.plot([-0.01,1], 
        [-0.01, 1], 
        'k--', label='y=x')
ax.plot(branch_lengths_LG, branch_lengths_LG_G4, 'o', label='branch lengths for LG and LG+G4 models')
 
# Labels and title
ax.set_xlabel('Branch lengths [LG]')
ax.set_ylabel('Branch lengths  [LG+G4] ')
ax.set_title('Branch lengths comparison between LG and LG+G4 [Codeml]')
ax.legend()

# point_x = posterior_mean_baseml[0]  # Example: first point in the list
# point_y = posterior_mean_iqtree[0]  # Example: first point in the list
# ax.annotate('Root age', xy=(point_x, point_y), xytext=(point_x - 0.1, point_y - 0.1), color='red')
# Show plot
# plt.show()
plt.savefig('./metazoa_branch_lengths_codeml_LG_G4_scatter.png')
```
