```python
import copy
import numpy as np
import matplotlib.pyplot as plt
```

```python
iqtree_file_LG = 'LG_G4_metazoa_in.BV'
codeml_file_LG = 'LG_G4_mcmctree_metazoa_out.BV'
```

```python
hessian_list = []

hessians_iqtree = []
treestrings_iqtree = []
branch_lengths_iqtree = []
gradient_vectors_iqtree = []
counter = 0
with open(iqtree_file_LG) as f:
    for k, line in enumerate(f.readlines()):
        if counter == 3:
            treestrings_iqtree.append(line)
            print('tree string',k)
        if counter == 5:
            branch_lengths_numerical = line.strip().split()
            branch_lengths_iqtree.append([float(i) for i in branch_lengths_numerical])
            print('branch lengths',k)
        if counter == 7:
            gradients_numerical = line.strip().split()
            gradient_vectors_iqtree.append([float(i) for i in gradients_numerical])
            print('gradients',k)
            
        if counter > 11 and line.strip() == '':
            hessians_iqtree.append(copy.deepcopy(hessian_list))
            hessian_list = []
            counter = 0 
            print('Hessian all string',k)
        elif counter > 11:
            # print(i, line)
            hessian_row_numerical = line.strip().split()
            hessian_list.append([float(i) for i in hessian_row_numerical])
            print('local hessian',k)
        counter += 1
    else:
        hessians_iqtree.append(copy.deepcopy(hessian_list))
```

```python
branch_lengths_iqtree = np.array(branch_lengths_iqtree)
```

```python
branch_lengths_iqtree.shape
```

```python
branch_lengths_iqtree = branch_lengths_iqtree.reshape(-1,1)
```

```python
branch_lengths_iqtree
```

```python
hessian_list_codeml = []

hessians_codeml= []
treestrings_codeml = []
branch_lengths_codeml = []
gradient_vectors_codeml = []
counter = 0
with open(codeml_file_LG) as f:
    for k, line in enumerate(f.readlines()):
        if counter == 3:
            treestrings_codeml.append(line)
            print('tree string',k)
        if counter == 5:
            branch_lengths_numerical = line.strip().split()
            branch_lengths_codeml.append([float(i) for i in branch_lengths_numerical])
            print('branch lengths',k)
        if counter == 7:
            gradients_numerical = line.strip().split()
            gradient_vectors_codeml.append([float(i) for i in gradients_numerical])
            print('gradients',k)
            
        if counter > 11 and line.strip() == '':
            hessians_codeml.append(copy.deepcopy(hessian_list_codeml))
            hessian_list_codeml = []
            counter = 0 
            print('Hessian all string',k)
        elif counter > 11:
            # print(i, line)
            hessian_row_numerical = line.strip().split()
            hessian_list_codeml.append([float(i) for i in hessian_row_numerical])
            print('local hessian',k)
        counter += 1
    else:
        hessians_codeml.append(copy.deepcopy(hessian_list_codeml))
        
```

```python
branch_lengths_codeml = np.array(branch_lengths_codeml)
```

```python
branch_lengths_codeml.shape
```

```python
branch_lengths_codeml = branch_lengths_codeml.reshape(-1,1)
```

```python
branch_lengths_codeml
```

```python
fig, ax = plt.subplots(figsize=(9 ,9))
ax.plot([-0.01,1], 
        [-0.01, 1], 
        'k--', label='y=x')
ax.plot(branch_lengths_codeml, branch_lengths_iqtree, 'o')
 
# Labels and title
ax.set_xlabel('Branch lengths-Codeml [LG+G4]')
ax.set_ylabel('Branch lengths-IQ-TREE [LG+G4] ')
ax.set_title('Branch lengths comparison between IQ-TREE and Codeml [LG+G4]')
ax.legend()

# point_x = posterior_mean_baseml[0]  # Example: first point in the list
# point_y = posterior_mean_iqtree[0]  # Example: first point in the list
# ax.annotate('Root age', xy=(point_x, point_y), xytext=(point_x - 0.1, point_y - 0.1), color='red')
# Show plot
# plt.show()
plt.savefig('./metazoa_branch_lengths_codeml_iqtree_scatter_LG_G4.png')
```
