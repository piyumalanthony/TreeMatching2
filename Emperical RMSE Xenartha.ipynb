### Code for IQ-TREE Hessian file

```python
hessian_list = []

hessians = []
treestrings = []
branch_lengths = []
gradient_vectors = []
counter = 0
with open(iqtree_file) as f:
    for k, line in enumerate(f.readlines()):
        if counter == 3:
            treestrings.append(line)
            print('tree string',k)
        if counter == 5:
            branch_lengths_numerical = line.strip().split()
            branch_lengths.append([float(i) for i in branch_lengths_numerical])
            print('branch lengths',k)
        if counter == 7:
            gradients_numerical = line.strip().split()
            gradient_vectors.append([float(i) for i in gradients_numerical])
            print('gradients',k)
            
        if counter > 11 and line.strip() == '':
            hessians.append(copy.deepcopy(hessian_list))
            hessian_list = []
            counter = 0 
            print('Hessian all string',k)
        elif counter > 11:
            # print(i, line)
            hessian_row_numerical = line.strip().split()
            hessian_list.append([float(i) for i in hessian_row_numerical])
            print('local hessian',k)
        counter += 1
    else:
        hessians.append(copy.deepcopy(hessian_list))
```

```python
def rmse(arr1, arr2):
    arr_1 = np.array(arr1)
    arr_2 = np.array(arr2)
    err = arr_1 - arr_2
    err_sqr = np.power(err, 2)
    err_sqr_mean = np.sum(err_sqr) / np.shape(err_sqr)[0]
    return np.sqrt(err_sqr_mean)
```

```python
branch_length_error = []
all_branches_iqtree = []
all_branches_mcmctree = []
for i in range(len(hessians_mcmctree)):
    branch_lengths_vector_iqtree = np.array(branch_lengths[i])
    branch_lengths_vector_mcmctree = np.array(branch_lengths_mcmctree[i])
    branch_length_error_data = rmse(branch_lengths_vector_iqtree, branch_lengths_vector_mcmctree)
    branch_length_error.append(branch_length_error_data)
    all_branches_iqtree.extend(branch_lengths[i])
    all_branches_mcmctree.extend(branch_lengths_mcmctree[i])
```
