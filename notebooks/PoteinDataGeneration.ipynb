```python
import os
import ete3
```

```python
file_str = "/home/piyumal/PHD/TimeTree/Protein_models/simulated_data"
IQTREE_PATH = '/home/piyumal/Tools/IQTREE/pointer_to_root/iqtree2'
```

```python
t = ete3.Tree()
taxa = ["T" + str(i) for i in range(1, 11)]
t.populate(10, taxa, random_branches=True, branch_range=[0.0005, 0.4])
t.write(format=1, outfile=f'{file_str}/tree.nw')
```

```python
cmd_alisim = f'{IQTREE_PATH} --alisim {file_str} -m WAG --seqtype AA -t {file_str}/tree.nw -seed 1 --length {10000}'
os.system(cmd_alisim)
```

```python
alignment_file = '/home/piyumal/PHD/TimeTree/Protein_models/simulated_data/simulated_data.phy'
species = []
sequences = []
summary = ''
with open(alignment_file, "r") as alignment:
    line_count = 0
    for line in alignment:
        if line_count==0:
            summary = line
        else:
            data = line.strip("\n").split()
            species.append(data[0])
            sequences.append(data[1])
        print(line)
        line_count += 1
print(line_count)
```
