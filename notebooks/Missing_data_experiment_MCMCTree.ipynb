```python
alignment_file = '/home/piyumal/PHD/TimeTree/Missing_data_experiment/Missing_data_experiment.phy'
species = []
sequences = []
summary = ''
with open(alignment_file, "r") as alignment:
    line_count = 0
    for line in alignment:
        if line_count==0:
            summary = line
        else:
            data = line.strip("\n").split()
            species.append(data[0])
            sequences.append(data[1])
        print(line)
        line_count += 1
print(line_count)
```
