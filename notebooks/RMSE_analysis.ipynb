```python
import numpy as np
```

```python
def rmse(arr1, arr2):
    arr_1 = np.array(arr1)
    arr_2 = np.array(arr2)
    err = arr_1 - arr_2
    err_sqr = np.power(err, 2)
    err_sqr_mean = np.sum(err_sqr) / np.shape(err_sqr)[0]
    return np.sqrt(err_sqr_mean)
```

```python
def read_files(file_path):
    Hessian = []
    branches = []
    gradients = []
    Hessian_diagonal = []
    Hessian_off_diagonal = []
    Hessian_all = []
    counter = 0
    with open(file_path, "r") as f:
        for line in f:
            if line.strip():
                counter+=1
                if counter == 3:
                    branches = [float(i) for i in line.strip().split()]
                if counter == 4:
                    gradients = [float(i) for i in line.strip().split()]
                if counter > 5:
                    Hessian.append([float(i) for i in line.strip().split()]) 
                print(line)
                print("##############################\n")
    for i,j in enumerate(Hessian):
        for l,m in enumerate(j):
            if i==l:
                Hessian_diagonal.append(m)
            else:
                Hessian_off_diagonal.append(m)
            Hessian_all.append(m)
    data = {"branches": branches, "gradients": gradients, "Hessian_diagonal":Hessian_diagonal, "Hessian_off_diagonal":Hessian_off_diagonal, "Hessian_all":Hessian_all}
    return data
        
```

```python
iqtree_file = "/home/piyumal/PHD/TimeTree/Protein_models/simulated_data/single_partition/output_v1.out.BV"
mcmc_file = "/home/piyumal/PHD/TimeTree/Protein_models/simulated_data/mcmctree_runs/single_partition/rst2"
```

```python
data1 = read_files(iqtree_file)
```

```python
data2 = read_files(mcmc_file)
```

```python
def load_rmse_data1(data1, data2):
    iqtree_br = np.array(data1["branches"])
    baseml_br = np.array(data2["branches"])
    error_br = rmse(iqtree_br, baseml_br)

    iqtree_gradients = np.array(data1["gradients"])
    baseml_gradients = np.array(data2["gradients"])
    error_gradients = rmse(iqtree_gradients, baseml_gradients)
    
    
    iqtree_off_diagonal_h = np.array(data1["Hessian_off_diagonal"])
    baseml_off_diagonal_h = np.array(data2["Hessian_off_diagonal"])
    
    iqtree_diagonal_h = np.array(data1["Hessian_diagonal"])
    baseml_diagonal_h = np.array(data2["Hessian_diagonal"])
    
    iqtree_h = np.array(data1["Hessian_all"])
    baseml_h = np.array(data2["Hessian_all"])
    min_h = min(iqtree_h)
    max_h = max(iqtree_h)

    error_diagonal_h = rmse(iqtree_diagonal_h, baseml_diagonal_h)
    error_off_diagonal_h = rmse(iqtree_off_diagonal_h, baseml_off_diagonal_h)
    error_h = rmse(iqtree_h, baseml_h)

    return error_br, error_gradients, error_diagonal_h, error_off_diagonal_h, error_h, min_h, max_h
```

```python
error_br, error_gradients, error_diagonal_h, error_off_diagonal_h, error_h, min_h, max_h = load_rmse_data1(data1, data2)
```
